Carilah kata-kata slangword yang ada dalam dataset Anda, dengan cara
mencocokan dengan kamus KBBI (terlampir). Simpan hasilnya dalam bentuk
csv/excel. Tampilkan 100 kata slang yang Anda dapatkan dan tampilkan dalam
bentuk Gunakan

```python
import pandas as pd

df_komentar = pd.read_csv('/content/komentar.csv')
df_slangword = pd.read_csv('/content/slangword.csv')

df_komentar['full_text'] = df_komentar['full_text'].str.lower()
df_slangword['Slangwords'] = df_slangword['Slangwords'].str.lower()

non_slang_found = []

for komentar in df_komentar['full_text']:
    for word in komentar.split():
        if word not in df_slangword['Slangwords'].values:
            non_slang_found.append(word)

df_non_slang_found = pd.DataFrame(non_slang_found[:100], columns=['full_text'])

df_non_slang_found.to_csv('non_slang_found.csv', index=False)

print(df_non_slang_found)
```

 Lakukan tokenizing berdasarkan  hasil 3b, simpan hasilnya dalam bentuk
csv/excel  dan tampilkan 100 token pertama

```python
df_slang_found = pd.read_csv('slang_found.csv')

tokens = []

for slangword in df_slang_found['full_text']:
    tokens.extend(word_tokenize(slangword))

df_tokens = pd.DataFrame(tokens[:100], columns=['token'])

df_tokens.to_csv('tokens_slang_words.csv', index=False)

print(df_tokens)
```

Lakukan stopword removing  berdasarkan hasil 3c, simpan hasilnya dalam
bentuk csv/excel

```python
pip install Sastrawi
```

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

df = pd.read_csv('tokens_slang_words.csv')

factory = StopWordRemoverFactory()
stopword = factory.create_stop_word_remover()

df['token'] = df['token'].apply(lambda x: stopword.remove(x))

df.to_csv('soal3d_hasilremoving.csv', index=False)
```

Lakukan stemming berdasarkan hasil 3d dan tampilkan 100 stem pertama.

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

df = pd.read_csv('soal3d_hasilremoving.csv')

factory = StemmerFactory()
stemmer = factory.create_stemmer()

df['token'] = df['token'].astype(str)
df['token'] = df['token'].apply(lambda x: stemmer.stem(x))

print(df['token'].head(100))

df.to_csv('soal3e_hasilstemming.csv', index=False)
```
