```python
import numpy as np
import pandas as pd
import os
from luwiji.text_proc import illustration
from nltk.tokenize import word_tokenize
from tqdm.auto import tqdm
from gensim.models import Word2Vec
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer, TfidfVectorizer
```

## Preprocessing

```python
# load dataset yang diberikan menjadi dataframe
df = pd.read_csv('E:\Kuliah 2021\Pelajaran\Praktikum NLP\Praktikum 5\data.csv')
df = df.drop(columns='Unnamed: 0')
df.head()
```

```python
import re
def remove_punctuation(oldtext):
    newtext = re.sub(r'[^A-Za-z]+', ' ', oldtext)
    return newtext

df['clean'] = df['content'].apply(remove_punctuation)
df['clean'] = df['clean'].str.lower()
df
```

```python
import nltk
nltk.download('stopwords')
```

```python
def filter_stopword(string:str):
    temp = string.split(' ')
    out=[]
    for w in temp:
        if w not in nltk.corpus.stopwords.words('indonesian'):
            #print(w)
            out.append(w)
    return ' '.join(out)
```

```python
df['clean'] = df['clean'].apply(filter_stopword)
df
```

## TF-IDF

```python
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(df['clean'])
print(tfidf_matrix)
```

```python
tfidf.get_feature_names_out()
```

```python
pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out(),
index=df['clean'])
```

```python
#Berikan penjelasan representasi output TFIDF dari sklearn
# Buat kumpulan dokumen
corpus = df['clean']

# Buat objek TfidfVectorizer
tfidf = TfidfVectorizer()

# Hitung tfidf dari kumpulan dokumen
tfidf_matrix = tfidf.fit_transform(corpus)

# Ubah matriks sparse menjadi tabel
tfidf_table = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out(), index=corpus)

# Tampilkan tabel
print(tfidf_table)
```
